# Decision Tree Classifier

    - Supervised learning algorithm
    - Decision trees are a non-parametric supervised learning method used for classification and regression. 
    - Decision tree implementation goal to create a model that predicts the value of a target variable by 
    learning simple decision rules inferred from the data features.

```python
from dataPrep import prepData
from visualize import plotIt
from sklearn import svm
```

```python
import matplotlib.pyplot as plt
import copy
import numpy as np
import pylab as pl
%matplotlib inline
```

```python
features_train, labels_train, features_test, labels_test = prepData()
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
```

```python
clf.fit(features_train, labels_train)
```

```python
pred = clf.predict(features_test)
```

```python
plotIt(clf, features_test, labels_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(pred, labels_test)

print (acc*100,'%')
```

# Checking by setting min_samples_split parameter

    min_samples_split or min_samples_leaf to ensure that multiple samples inform every decision in the tree, by
    controlling which splits will be considered. 
    
    A very small number will usually mean the tree will overfit, whereas a large number will prevent the tree 
    from learning the data.
     
    Min value of min_samples_split can be 2 as after that the node can't be split into 2
    
###### Setting min_samples_split = 50 and checking accuracy

###### for this data the highest accuracy comes with min_samples_split = 19

```python
clf = tree.DecisionTreeClassifier(min_samples_split=50)
clf.fit(features_train, labels_train)
```

```python
pred = clf.predict(features_test)
```

```python
plotIt(clf, features_test, labels_test)
```
