# Decision Tree Classifier

#### Definition
    - Supervised learning algorithm
    - Decision trees are a non-parametric supervised learning method used for classification and regression. 
    - Decision tree implementation goal to create a model that predicts the value of a target variable by 
    learning simple decision rules inferred from the data features.

## How Decision tree creates Non-Linear decision boundary from Linear decision surfaces

Like SVM kernel trick, which helps to change decision boundary from linear to non-linear decision surfaces, Decision Trees use a trick to do non-linear decision making with simple decision surfaces

# Scatterplot Explanation from Notes

# Entropy 
### How a decision tree decides where to split the data

Its a measure of impurity in a bunch of examples.

# Scatterplot explanations

When you build a decision tree, you try to find variables and split points along those variables that's going to make subsets that are as pure as possible. 

###### And the above process is repeated recursively and that's how decision tree makes its decision.

# Formula of entropy
\begin{equation*}
\left( \sum_{i=0}^n p_i    log_2 p_i\right)
\end{equation*}

#### p_i - Fraction of examples in class i

```python
from dataPrep import prepData
from visualize import plotIt
from sklearn import svm
```

```python
import matplotlib.pyplot as plt
import copy
import numpy as np
import pylab as pl
%matplotlib inline
```

```python
features_train, labels_train, features_test, labels_test = prepData()
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
```

```python
clf.fit(features_train, labels_train)
```

```python
pred = clf.predict(features_test)
```

```python
plotIt(clf, features_test, labels_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(pred, labels_test)

print (acc*100,'%')
```

# Checking by setting min_samples_split parameter

    min_samples_split or min_samples_leaf to ensure that multiple samples inform every decision in the tree, by
    controlling which splits will be considered. 
    
    A very small number will usually mean the tree will overfit, whereas a large number will prevent the tree 
    from learning the data.
     
    Min value of min_samples_split can be 2 as after that the node can't be split into 2
    
###### Setting min_samples_split = 50 and checking accuracy

###### for this data the highest accuracy comes with min_samples_split = 19

```python
#clf = tree.DecisionTreeClassifier(min_samples_split=50)
clf = tree.DecisionTreeClassifier(min_samples_split=19)
clf.fit(features_train, labels_train)
```

```python
pred = clf.predict(features_test)
```

```python
plotIt(clf, features_test, labels_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(pred, labels_test)

print (acc*100,'%')
```
