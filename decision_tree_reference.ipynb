# Decision Tree Classifier

    - Supervised learning algorithm
    - Decision trees are a non-parametric supervised learning method used for classification and regression. 
    - Decision tree implementation goal to create a model that predicts the value of a target variable by 
    learning simple decision rules inferred from the data features.

```python
from dataPrep import prepData
from visualize import plotIt
from sklearn import svm
```

```python
import matplotlib.pyplot as plt
import copy
import numpy as np
import pylab as pl
%matplotlib inline
```

```python
features_train, labels_train, features_test, labels_test = prepData()
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
```

```python
clf.fit(features_train, labels_train)
pred = clf.predict(features_test)
```

```python
plotIt(clf, features_test, labels_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(pred, labels_test)

print (acc*100,'%')
```
